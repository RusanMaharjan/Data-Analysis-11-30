# Feature Engineering

```python
df['DOB'].max()
```

```python
df['Birth_Year'] = df['DOB'].dt.year
```

```python
df['Birth_Month'] = df['DOB'].dt.month
df['Birth_Day'] = df['DOB'].dt.day
```

```python
df['Birth_Month_Name'] = df['DOB'].dt.month_name()
df['Birth_Day_Name'] = df['DOB'].dt.day_name()
```

```python
# split(find, n(number of columns)=index, expand=False(default)/True)
# name = df['Employee_Name'].str.split(' ', n=2, expand=True)
# name
```

```python
last_first = df['Employee_Name'].str.split(',', n=1, expand=True)
last_first
```

```python
first_middle = last_first[1].str.strip().str.split(' ', n=1, expand=True)
first_middle
```

```python
df['First_Name'] = first_middle[0]
df['Middle_Name'] = first_middle[1]
df['Last_Name'] = last_first[0]
```
