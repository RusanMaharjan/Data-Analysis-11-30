## Conditional Filtering

```python
import pandas as pd
```

```python
df = pd.read_csv('Dataset/Employees.csv')
df.head()
```

```python
# Find details of employees who earns more than 11000.
# select * from employees where salary > 11000;
emp_sal = df['SALARY'] > 11000
df[emp_sal]
```

```python
emp_sal = df[df['SALARY'] > 11000]
emp_sal
```

```python
# Find details of employees who works in department 50 and earns more than 5000.
dep_sal = df[(df['DEPARTMENT_ID'] == 50) & (df['SALARY'] > 5000)]
dep_sal
```

```python
# Find details of employees who works either in job SA_MAN or IT_PROG.
emp_job = df[(df['JOB_ID'] == 'SA_MAN') | (df['JOB_ID'] == 'IT_PROG')]
emp_job
```

```python
# Find details of employees whose salary is in range of 1000 to 5000.
emp_sal_bet = df[df['SALARY'].between(1000, 5000)]
emp_sal_bet
```

```python
# Find details of employees who works either in job SA_MAN or IT_PROG.
emp_job_or = df[df['JOB_ID'].isin(['SA_MAN', 'IT_PROG'])]
emp_job_or[['FIRST_NAME', 'LAST_NAME', 'JOB_ID', 'SALARY', 'DEPARTMENT_ID']]
```

```python
# NaN -> Not a Number
# NaT -> Not a Time
```

```python
df = pd.read_csv('Dataset/Employees.csv')
df.head()
```

```python
'''
dropna(axis=0(default)/1, how='any' (default)/'all', inplace=True/False(default))
--------------------
axis
    -> 0 -> Row
    -> 1 -> Column

how
    -> any -> one or many null values
    -> all -> all data must be null
'''
```

```python
# Inplace=True -> Make changes permanently
df.dropna(axis=1, how='any', inplace=True)
```

```python
df
```

```python
df.columns
```

```python
# fillna(value, inplace=True/False(default))
df.isnull().sum()
```

```python
mean_comm = df['COMMISSION_PCT'].mean()
median_comm = df['COMMISSION_PCT'].median()

print(mean_comm)
print(median_comm)
```

```python
# df['COMMISSION_PCT'].fillna(median_comm, inplace=True)
df['COMMISSION_PCT'] = df['COMMISSION_PCT'].fillna(median_comm)
```

```python
df['COMMISSION_PCT'].unique()
```

```python
mgr_mode = df['MANAGER_ID'].mode()[0]
```

```python
dep_mode = df['DEPARTMENT_ID'].mode()[0]
dep_mode
```

```python
df['MANAGER_ID'] = df['MANAGER_ID'].fillna(mgr_mode)
```

```python
df['DEPARTMENT_ID'] = df['DEPARTMENT_ID'].fillna(dep_mode)
```

```python
df.info()
```

### Change datatypes

```python
df['MANAGER_ID'] = df['MANAGER_ID'].astype(int)
```

```python
df['DEPARTMENT_ID'] = df['DEPARTMENT_ID'].astype(int)
```

```python
'''
DateTime
------------
2/21/2001
2001/04/23
2003-04-5

Format
---------
'%d.%m.%y'
'%d-%m-%y'
'mixed'
'''
```

```python
df['HIRE_DATE'] = pd.to_datetime(df['HIRE_DATE'], format='%d.%m.%y')
```

### Skewness and Kurtosis

```python
## Skewness
'''
    postive skewness -> right skewness -> number of low values are more
    negative skewness -> left skewness -> number of high values are more
    zero -> normal distribution -> high and low value both are equal
'''
```

```python
df['SALARY'].skew()
```

```python
'''
High Kurtosis -> High Outlier
Low Kurtosis -> Flatter distribution
'''
```

```python
df['SALARY'].kurt()
```
