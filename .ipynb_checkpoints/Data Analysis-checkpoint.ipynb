## Group By

```python
'''
1. if we use Group by then there must be use of aggregate function.
2. If we only used aggregate function then don't need to use group by function.
'''
```

### Find total salary by each job id

```python
job_emp = df.groupby('JOB_ID')['SALARY'].sum()
job_emp
```

```python
df['SALARY'].sum()
```

```python
df['EMPLOYEE_ID'].count()
```

### Print data using reset function, rename columns, sorting values

```python
job_emp = df.groupby('JOB_ID')['SALARY'].sum().reset_index().rename(
    columns = {'JOB_ID': 'Job Name', 'SALARY': 'Total Salary'}
).sort_values(ascending=False, by='Total Salary', ignore_index=True)
job_emp
```

### Prints data uisng markdown

```python
job_emp = df.groupby('JOB_ID')['SALARY'].sum().sort_values(ascending=False)
format_data = job_emp.map("Rs.{:,.2f}".format).to_markdown(numalign='right', stralign='center')
print(format_data)
```

### Find total employees in each departments.

```python
dep_emp = df.groupby('DEPARTMENT_ID')['EMPLOYEE_ID'].count().reset_index()
dep_emp
```

### Filter data using agg function and automation currency converting

```python
dep_emp_sal = df.groupby('DEPARTMENT_ID').agg(
    Total_Employees = ('EMPLOYEE_ID', 'count'),
    Total_Salary = ('SALARY', 'sum'),
    Max_Salary = ('SALARY', 'max'),
    Min_Salary = ('SALARY', 'min'),
    Average_Salary = ('SALARY', 'mean')
).reset_index()
# dep_emp_sal
```

```python
def currencyConverter(df_name, column_name):
    for col in column_name:
        df_name[col] = df_name[col].map("Rs.{:,.2f}".format)
```

```python
dep_emp_sal = df.groupby('DEPARTMENT_ID')['SALARY'].agg(
    Total_Salary = ('sum'),
    Max_Salary = ('max'),
    Min_Salary = ('min'),
    Average_Salary = ('mean')
).reset_index()
# dep_emp_sal
```

```python
columns = ['Total_Salary', 'Max_Salary', 'Min_Salary', 'Average_Salary']
```

```python
currencyConverter(dep_emp_sal, columns)
```
